```python

import matplotlib.pyplot  as plt
import numpy as np
 

plt.figure(figsize=(20,5))
t= np.linspace(0,20,100)
y= np.sin(t)
x= np.cos(t)
plt.plot(t,y, label='sin(t)', color='orange', linewidth=3, marker ='o',markersize=10)
plt.plot(t,x, label='cos(t)', color='hotpink', linewidth=2, linestyle='--')
plt.text(10, 0.5, 'sin(t)', fontsize=20)
plt.text(10,-0.5,'cos(t)', fontsize=20)
plt.legend()
plt.grid(False)
```
